# Using figure objects

```python
# A figure is an object that contains all the plot elements
# It can contain many axes
# Define width & height in inches
# Dots per inch
fig_1 =plt.figure(figsize=(6,5),dpi=100)

# Adds axes with a left, bottom, width and height that ranges from 0 to 1
# which is the percent of the canvas you want to use
axis_1=fig_1.add_axes([0.1,0.1,0.9,0.9])

# add labels and title
axis_1.set_title("Days square chat",size=25)
axis_1.set_xlabel("Days")
axis_1.set_ylabel("Days square")

# Plot on the axes (If you want a label associated with the legend add it with label)
axis_1.plot(x_1,y_1,label="x/x2")

# You can plot to plots using the same axes
axis_1.plot(y_1,x_1,label="x2/x")

axis_1.legend(loc="upper right")

#functions are used to set the x-axis and y-axis view limits 
axis_1.set_xlim([0,40])
axis_1.set_ylim([0,40])

# Add the optional legend with a location number (best: 0,upper right: 1, upper left: 2, lower left: 3, lower right: 4,
# or supply a tuple of x & y from lower left
axis_1.legend(loc=0)

# You can create graph inside of other graph
axis_2 =fig_1.add_axes([0.4,0.4,0.4,0.35])
axis_2.set_title("Days square chat",size=15)
axis_2.set_xlabel("Days")
axis_2.set_ylabel("Days square")
axis_2.plot(x_1,y_1,'r')
axis_2.set_xlim([0,15])
axis_2.set_ylim([0,110])

# add text message in graph
axis_2.text(1,150,"Add Message")

''''# figure 2
fig_2 =plt.figure(figsize=(6,5),dpi=100)
axis_3=fig_2.add_axes([0.1,0.1,0.9,0.9])
axis_3.set_title("Days square chat")
axis_3.set_xlabel("Days")
axis_3.set_ylabel("Days square")
axis_3.plot(x_1,y_1,label="x/x2")
axis_3.plot(y_1,x_1,label="x2/x")

axis_3.legend(loc=0)

axis_4 =fig_2.add_axes([0.4,0.4,0.4,0.35])
axis_4.set_title("Days square chat")
axis_4.set_xlabel("Days")
axis_4.set_ylabel("Days square")
axis_4.plot(x_1,y_1,label="x/x2")'''

# improve the space between subplot and the axis labels and titles are not cut off
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np 

x_1 = np.linspace(1,5,5)
y_1 = x_1+2
fig1 = plt.figure(figsize=(6,5))
axis1 = fig1.add_axes([0,0,0.5,0.5])
axis1.plot(x_1,y_1,'k',alpha=0.99,lw=1,ls="solid",marker="s",markersize="6",markerfacecolor="k",markeredgecolor='r',markeredgewidth='1')
axis1.set_title("Graph",size=25)
axis1.set_xlabel("X-axis")
axis1.set_ylabel("Y-axis")
axis1.set_xlim([0,6])
axis1.set_ylim([2,8])

# Add a grid true print normal graph lines in background
# color is thickness of the graph lines in background 
# dashes (5pts 1 pt dashes separated by 2pt space) 
axis1.grid(True,color='0.1',dashes=(5,2,1,2))

# Set grid background colorf
axis1.set_facecolor('#FAEBD7')

plt.savefig("sss.jpg")
```
